```python
import numpy as np  # Calcul numérique
import pandas as pd  # Analyse et manipulation de données
import matplotlib.pyplot as plt  # Visualisation des données sous forme de graphes et courbes
import seaborn as sns  # Création rapide de graphiques attractifs
```

```python
#Importer LinearRegression de scikit-learn pour appliquer la régression linéaire
from sklearn.linear_model import LinearRegression

#Importer train_test_split de scikit-learn pour diviser les données en ensembles d'entraînement et de test
from sklearn.model_selection import train_test_split

```

```python
#Charger le fichier CSV dans un DataFrame en utilisant la tabulation comme séparateur
data = pd.read_csv('/content/drive/My Drive/Salary_dataset.csv', sep = ",")
# Afficher data
data
```

```python
data.head()
```

```python
data.info()
```

```python
# Afficher un graphique en nuage de points
X = data["YearsExperience"]
Y = data["Salary"]
plt.scatter(X, Y, color='red')
```

```python
X = data['YearsExperience']
Y = data['Salary']
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2)
```

```python
# Créer un graphique en nuage de points pour visualiser la relation entre X_train et Y_train
plt.scatter(X_train, Y_train)
plt.xlabel("X_train")
plt.ylabel("Y_train")
plt.show()# Affiche le graphique à l'écran
```

```python
# Créer une instance du modèle de régression linéaire
model = LinearRegression()

# Ajuster (entraîner) le modèle avec les données d'entraînement
model.fit(X_train.values.reshape(-1,1), Y_train.values.reshape(-1,1))
```

```python
prediction = model.predict(X_test.values.reshape(-1,1))
#affichage du résultat
prediction
```

```python
plt.scatter(X_test, prediction, color='red')
plt.plot(X_train.values.reshape(-1,1), model.predict(X_train.values.reshape(-1,1)), color='blue')
plt.xlabel("X_test")
plt.ylabel("Y_test")
plt.show()
```

```python
score = model.score(X_test.values.reshape(-1,1), Y_test.values.reshape(-1,1))
print("Score = ", score)
```

```python
from sklearn.metrics import mean_squared_error
# Prédictions du modèle sur les données de test
y_pred = model.predict(X_test.values.reshape(-1, 1))
# Calcul du MSE
mse = mean_squared_error(Y_test.values.reshape(-1, 1), y_pred)
print("Mean Squared Error = ", mse)
```
